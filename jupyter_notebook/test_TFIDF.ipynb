```python
import jieba
import numpy as np
import copy
## 載入自訂字典
KeyWords_path = "./keyWordsDB"
jieba.load_userdict(os.path.join(KeyWords_path, "AllKeyWords.txt"))
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import operator
```

```python
def get_sgm_text(txt_file_path):
    text = ""
    with open(txt_file_path, "r" , encoding="utf8")as txt:
        for line in txt.readlines():
            for sgm in jieba.cut(line.strip()):
                #print(sgm.strip())
                if sgm.strip() != "": 
                    text = text + sgm.strip() + " "
    return text
```

```python
def get_tfidf_table(sgm_txt_list):
    ## 將count.vocabulary_的字典資料做排序
    def sort_dict_by_values(count):
        return  sorted(count.vocabulary_.items(), key = operator.itemgetter(1))
    
    count = CountVectorizer()
    bag = count.fit_transform(sgm_txt_list)
    tfidf = TfidfTransformer()
    tfidf_result = tfidf.fit_transform(bag)
    
    ##  產生這次TFIDF分析後的DF資料
    df = pd.DataFrame(tfidf_result.toarray(), columns= [x[0] for x in sort_dict_by_values(count)])
    return df
```

```python
rawdata_path = "./chemicals_rawdata/chemicals_txt"
```

```python
sgm_txt_list = []
for txt_file in os.listdir(rawdata_path):
    sgm_txt_list.append(get_sgm_text(os.path.join(rawdata_path, txt_file)))
```

```python
tfidf_df = get_tfidf_table(sgm_txt_list)
tfidf_df.transpose().to_excel("TFIDF_result.xlsx")
```

```python
t_tfidf_df = tfidf_df.transpose()
```

```python
def isin_chinese(txt):
    have_chinese = False
    for s in txt:
        if len(s.encode("utf-8"))>1:
            have_chinese = True
    return have_chinese
```

```python
q1 = t_tfidf_df.mean(1).quantile(0.25)
q3 = t_tfidf_df.mean(1).quantile(0.75)

for idx in t_tfidf_df[t_tfidf_df.mean(1) >q3].index:
    if isin_chinese(idx):
        print(idx)
```
